```python
from imputime import tempute
import pandas as pd
```

```python
months = pd.read_excel('example_data.xlsx',sheet_name = 'month')
```

Below we can see that this data is at the monthly level and there are gaps

```python
months
```

We confirm this with the gap_check function

```python
tempute.gap_check(months, 'months')
```

We can impute the missing values using the fill_gaps function. In this case we will assume a linear rate of change.

```python
months_filled = tempute.fill_gaps(months, 'linear', 'value', 'months')
months_filled
```

Now that the gaps in the data are filled, we can turn the data into a daily interval

```python
daily = tempute.monthly_to_daily(months_filled, 'linear', 'value')
daily.head(31)
```

We can also extrapolate the data from any time point. In this example I will extrapolate from 9/27/2024 for 5 days using the average rate of change of the last 5 days.

```python
tempute.extrapolate(daily, 'linear', 'value', 'days', '2024-09-27', 5, -5)
```
